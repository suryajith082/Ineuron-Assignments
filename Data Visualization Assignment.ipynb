In this assignment students have to transform iris data into 3 dimensions
and plot a 3d chart with transformed dimensions and colour each data
point with specific class.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
iris = datasets.load_iris()
x = iris.data
y = iris.target
```

```python
fig = plt.figure(1,figsize = (11,7)) 
m = Axes3D(fig, elev=-150, azim=120)  
m.scatter(x[:,0],x[:,1],x[:,2], c = y, cmap = plt.cm.rainbow_r)
m.set_title("IRIS Dataset in 3d scatter plot")
m.set_xlabel("X data",labelpad=20)
m.set_ylabel("Y data",labelpad=20)
m.set_zlabel("Z data",labelpad=10)
m.set
plt.show()
```
